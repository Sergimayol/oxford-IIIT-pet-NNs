```python
import cv2
from torchvision import transforms
import os
```

```python
trimap_path = os.path.join(os.getcwd(), "..", "data", "annotations", "trimaps", "Abyssinian_1.png")

img_gray = cv2.imread(trimap_path, cv2.IMREAD_GRAYSCALE)
img_rgb = cv2.imread(trimap_path, cv2.IMREAD_COLOR)
```

```python
to_tensor = transforms.ToTensor()
tt_gray = to_tensor(img_gray)
tt_rgb = to_tensor(img_rgb)

tt_gray.shape, tt_rgb.shape
```

```python
# tt_gray.tolist()
tmp = tt_gray * 255
#tmp.tolist()
```

```python
import matplotlib.pyplot as plt

# Plot the image, permute the channels to get the color right
plt.imshow(tt_gray.permute(1, 2, 0))
```

```python
# Remove all the 1s
tmp_without_1 = tmp.clone()
tmp_without_1[tmp_without_1 == 1] = 0
print(tmp_without_1.unique())
plt.imshow(tmp_without_1.permute(1, 2, 0), cmap='gray')
```

```python
# Remove all the 1s
tmp_without_2 = tmp.clone()
tmp_without_2[tmp_without_2 == 2] = 0
plt.imshow(tmp_without_2.permute(1, 2, 0), cmap='gray')
```

```python
# Remove all the 3s
tmp_without_3 = tmp.clone()
tmp_without_3[tmp_without_3 == 3] = 0
plt.imshow(tmp_without_3.permute(1, 2, 0), cmap='gray')
```

```python
tmp_without_1_2 = tmp.clone()
tmp_without_1_2[tmp_without_1_2 == 1] = 1
tmp_without_1_2[tmp_without_1_2 == 2] = 0
tmp_without_1_2[tmp_without_1_2 == 3] = 1
plt.imshow(tmp_without_1_2.permute(1, 2, 0), cmap="gray")
print(tmp_without_1_2.shape)
```
